# 데이터의 비율을 표현하는 파이차트
- 파이차트를 이용하여 데이터의 상대적 비율을 표현할 수 있다.
___

```python
import matplotlib.pyplot as plt
import pandas as pd

# 그래프에 한글 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

## 1. 파이차트
- 전체에 대한 각 부분의 비율을 부채꼴 모양으로 나타낸 그래프이다.
- 각 부채꼴의 중심각이 전체에서 해당 데이터가 차지하는 비율을 나타낸다.
- 1차원 리스트/배열/시리즈를 이용하여 그린다.

### 1.1. 샘플 데이터


```python
# 2019년 병역판정검사 - 혈액형 분포
blood_type = ['A', 'B', 'O', 'AB']
Personnel = [111901, 87066, 86804,36495]
```

### 1.2. 기본 파이그래프 그리기
- plt.pie(data)

```python
plt.pie(Personnel)

plt.show()
```

### 1.3. 레이블 달기
- labels = [label list]
- labeldistance = 그래프로부터 레이블을 얼마나 떨어뜨려 표시할 것인가(default = 1.1)
 -> 그래프의 테두리가 1이다. 따라서, 1보다 작게 설정하면 레이블을 그래프 안으로 삽입 가능.

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5)
plt.show()
```

#### 1.4. 비율 표시하기
- autopct = '% . 소수점자리수 f %%' f는 float형을 의미
- pctdistance = 중심에서의 거리(반지름을 1이라 했을 때 반지름으로부터 얼마나 떨어져 비율을 표시할 지 정함)


```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8)
plt.show()
```

#### 1.5. 돌출효과
- explode = [돌출정도 리스트]
- 반지름의 길이를 1이라고 했을 때를 기준(default : 0)

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0.5,0.5,0.5,0.5])
plt.show()
```

#### 1.6. 색상 바꾸기
- colors = [color list]


```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0.5,0.5,0.5,0.5]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue'])
plt.show()
```

#### 1.7. 시작 각도
- startangle = 시작각도
- 기본 시작 각도는 3시방향으로, startangle을 지정하면 반 시계방향으로 각도만큼 이동하여 시작한다.

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0.5,0.5,0.5,0.5]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90) # 90도만큼 반시계 방향으로 회전됨
plt.show()
```

#### 1.8 회전방향
- countercolck = True / False (반시계, 시계)
- default = True

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0.5,0.5,0.5,0.5]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90, counterclock=False) # 90도만큼 시계 방향으로 회전됨
plt.show()
```

#### 1.9. 범례
- legend(레이블리스트)

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0,0,0,0]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90, counterclock=False)
plt.legend(loc=(1,0.8)) # loc 파라미터로 범례의 위치 지정
plt.show()
```

#### 1.10 반지름
- radius = 반지름(default : 1)

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0,0,0,0]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90, counterclock=False, radius=0.5)
plt.legend(loc=(1,0.8))
plt.show()
```

#### 1.11 부채꼴 스타일
- wedgeprops = {'ec':테두리컬러, 'lw':선두께, 'ls':선스타일, 'width':반지름과의 비율로 부채꼴의 형태가 도넛형으로 바뀐다}

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0,0,0,0]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90, counterclock=False, radius=1.5
, wedgeprops={'ec':'k', 'lw':2, 'ls': ':', 'width': 0.8}
) # lw : 1이 기준
plt.legend(loc=(1,0.8))
plt.show()
```

#### 1.12 폰트
- textprops = {'fontsize':폰트사이즈, 'color':폰트컬러, 'rotation':폰트회전}
- rotation의 경우, 반시계 방향으로 회전시킨다.

```python
plt.pie(Personnel, labels=blood_type, labeldistance=0.5,
autopct='%.2f%%', pctdistance=0.8, explode=[0,0,0,0]
, colors=['lightcoral', 'gold','greenyellow', 'skyblue']
, startangle=90, counterclock=False, radius=1.5
, wedgeprops={'ec':'k', 'lw':2, 'ls': ':', 'width': 0.8}
, textprops= {'fontsize': 15, 'color':'r', 'rotation':-45}
) # lw : 1이 기준
plt.legend(loc=(1,0.8))
plt.show()
```
